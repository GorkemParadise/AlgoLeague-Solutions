```python
N, C = map(int, input().split())
M = list(map(int, input().split()))

if C >= sum(M):
    print("EV")
else:
    for i in range(N):
        k = C - M[i]
        if k >= 0:
            C = k
        else:
            print(i + 1)
            break
```

```python
N, C = map(int, input().split())
M = list(map(int, input().split()))

if C >= sum(M):
    print("EV")
else:
    for i in range(N):
        if C >= M[i]:
            C -= M[i]
        else:
            print(i + 1)
            break
```

```python
N, C = map(int, input().split())
M = list(map(int, input().split()))

# Canavarları küçükten büyüğe sırala
M.sort()

for i in range(N):
    if C >= M[i]:
        C -= M[i]
    else:
        print(i + 1)  
        break
else:
    print("EV")      
```

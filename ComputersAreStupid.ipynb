```python
Q = int(input().strip())

for _ in range(Q):
    a, b = map(int, input().strip().split())
    n = a * b
    primes = set()
    d = 2
    while d * d <= n:
        while n % d == 0:
            primes.add(d)
            n //= d
        d += 1
    if n > 1:
        primes.add(n)
    print(len(primes))
```

```python
def prime_factors(x):
    primes = set()
    
    # Check for factor 2
    if x % 2 == 0:
        primes.add(2)
        while x % 2 == 0:
            x //= 2
    
    # Check odd factors from 3 onwards
    d = 3
    while d * d <= x:
        if x % d == 0:
            primes.add(d)
            while x % d == 0:
                x //= d
        d += 2
    
    if x > 1:
        primes.add(x)
    return primes

Q = int(input().strip())
for _ in range(Q):
    a, b = map(int, input().strip().split())
    primes = prime_factors(a) | prime_factors(b)
    print(len(primes))
```

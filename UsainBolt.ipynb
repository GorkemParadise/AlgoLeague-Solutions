```python
a = int(input().strip()) #rekor sayısı - 5
b = [int(x) for x in input().split()] #rekorlar - 1 2 3 4 5

my_list = [b[0]]
for i in range (1, a):
    if b[i] > b[i-1]:
        my_list.append(b[i])

print(len(my_list)) #output 4
```

```python
a = int(input().strip()) #rekor sayısı - 5
b = [int(x) for x in input().split()] #rekorlar - 3 7 2 6 8 

my_list = [b[0]]
for i in range (1, a):
    if b[i] > b[i-1]:
        my_list.append(b[i])

print(len(my_list)-1)  #output - 2
```

```python
a = int(input().strip())  #rekor sayısı - 5
b = list(map(int, input().split()))  #rekorlar - 3 7 2 6 8 

high = b[0]
high_breaks = 0

for i in range(1, a):
    if b[i] > high:
        high = b[i]
        high_breaks += 1

print(high_breaks)  #output - 2
```

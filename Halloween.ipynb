```python
M = int(input())
N = list(map(int, input().strip().split()))
A = int(input())

k = sum(N) - max(N)
l = A - k
if l % M == 0:
    print("1")
else:
    print("0")
```

```python
M = int(input())
N = list(map(int, input().split()))
A = int(input())

toplam = sum(N) + A

if toplam % M != 0:
    print(0)
else:
    hedef = toplam // M
    if hedef >= max(N):
        print(1)
    else:
        print(0)
```

```python
M = int(input())
Ai = list(map(int, input().split()))
N = int(input())

t = sum(max(Ai) - a for a in Ai)
print(1 if N >= t and (N - t) % M == 0 else 0)
```

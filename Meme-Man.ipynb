```python
N = int(input())
x, y = map(int, input().split())
```

```python
dp = set()
dp = [0] * (N + 1)
dp[0] = 1
for i in range(1, N + 1):
    if i - x >= 0:
        dp[i] += dp[i - x]
    if i - y >= 0:
        dp[i] += dp[i - y]
print(dp[N])
```

```python
import sys
import bisect
input = sys.stdin.readline
N, Q = map(int, input().split())
R = [tuple(map(int, input().split())) for _ in range(N)]
R.sort()
LL = [r[0] for r in R]
RR = [r[1] for r in R]
for _ in range(Q):
    x = int(input())
    idx = bisect.bisect_right(LL, x) - 1
    if idx >= 0 and RR[idx] >= x:
        print("Yes")
    else:
        print("No")
```

```python
def binary_search_right(arr, x):
    left, right = 0, len(arr)
    while left < right:
        mid = (left + right) // 2
        if arr[mid] <= x:
            left= mid + 1
        else:
            right = mid
    return left 

N, Q = map(int, input().split())
R = [tuple(map(int, input().split())) for _ in range(N)]
R.sort()

LL = [r[0] for r in R]
RR = [r[1] for r in R]

for _ in range(Q):
    x = int(input())

    idx = binary_search_right(LL, x) - 1

    if idx >= 0 and RR[idx] >= x:
        print("Yes")
    else:
        print("No")
```

```python
def generate_binary_strings(n):
    result = []
    def backtrack(current):
        if len(current) == n:
            result.append(current)
            return
        backtrack(current + '0')
        backtrack(current + '1')
    backtrack('')
    return result
```

```python
generate_binary_strings(2)
```

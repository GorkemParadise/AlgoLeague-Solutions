```python
n, k = map(int, input().split())

# 1. Hangi satırda olduğunu bulalım
# ileri: 1 -> 2 -> ... -> n, sonra geri: n-1 -> ... -> 1 ve tekrar
pos = 1          # başlangıç satırı
direction = 1    # 1: ileri, -1: geri

for _ in range(k - 1):  # ilk sıçrama zaten 1. satır
    if pos == n:
        direction = -1
    elif pos == 1:
        direction = 1
    pos += direction

# 2. Satırdaki kare numaralarını bul
current = 1
for row in range(1, pos):
    if row % 2 == 1:
        current += 1
    else:
        current += 2

# 3. Son satırın karelerini yazdır
if pos % 2 == 1:
    print(current)
else:
    print(current, current + 1)
```

```python
def fehmi_hopscotch(n, k):
    hopscotch = []
    for i in range(1, n + 1):
        if i % 2 == 1: 
            hopscotch.append([i])
        else:
            hopscotch.append([i * 2 - 1, i * 2])

    
    row = 0
    direction = 1  
    for _ in range(k):
        row += direction
        if row == n:  # GERİ
            direction = -1
        elif row == 0:  # DÖN
            direction = 1

    print(hopscotch[row])

n = 6  # rows
k = 20  # jumps
fehmi_hopscotch(n, k)
```

```python
n = 6  # rows
k = 20  # jumps

lst = []
cnt = 1
for i in range(n):
    if i % 2 == 0:
        lst.append(cnt)
        cnt += 1
    else:
        lst.append([cnt, cnt+1])
        cnt += 2

# İleri-geri hareketi hesaplama
cycle = 2*n - 2  # 1->n->1->n arası bir tam tur
jump_pos = k % cycle
if jump_pos == 0:
    jump_pos = cycle

if jump_pos <= n:
    row = jump_pos
else:
    row = n - (jump_pos - n)

# Son satırdaki kareleri yazdır
print(lst[row-1])
```

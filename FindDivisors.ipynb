```python
n = int(input())
yoo = set()
for i in range(int(n**(1/2))):
    if n % (i + 1) == 0:
        yoo.add(i + 1)
        yoo.add(n // (i + 1))
print(len(yoo))
print(*sorted(yoo))
```

```python
n = int(input())
divs = set()

for i in range(1, int(n**0.5) + 1):
    if n % i == 0:
        divs.add(i)
        divs.add(n // i)

print(len(divs))
print(*sorted(divs))
```

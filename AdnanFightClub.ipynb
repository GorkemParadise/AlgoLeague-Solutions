```python
Q = int(input().strip())
adnan = set()

for _ in range(Q):
    n = list(map(int, input().strip().split()))
    t = n[0]

    if t == 1:
        a = n[1]
        adnan.add(a)
    else:
        a = n[1]
        b = n[2]
        if (a-b in adnan) and (a+b in adnan) and (a in adnan):
            print("GG EZ")
        else:
            print("GLHF")

'''
4
1 5
1 3
1 7
2 5 2
'''
# GG EZ
```
